```python
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import roc_auc_score, average_precision_score, brier_score_loss

# -------- 0) Carga y features --------
# df = ... (tipificado como ya definimos)
df = df[df['edad_(años_cumplidos)'].between(30,74)].copy()

# Etiqueta (ejemplo: alto riesgo si clasificacion_final_del_riesgo == 'Alto')
# Si no existe, reemplaza 'label' por la etiqueta derivada de Framingham.
label_col = 'label_alto_rcv'  # crea esta columna antes
# df[label_col] = (df['clasificación_final_del_riesgo'].str.contains('Alto', case=False, na=False)).astype(int)

# Features (sin lípidos, cobertura)
df['imc'] = df['23_peso'] / (df['24_talla']**2)
df['hta_tratada'] = df[['36_rcbeieca','37_rcbeara2']].any(axis=1).astype(int)

X = df[['edad_(años_cumplidos)','8_sexo','25_tenartsis','imc','cigarrillo','20_dxdm','hta_tratada']].copy()
y = df[label_col].astype(int)

# Separar tipos
num_cols = ['edad_(años_cumplidos)','25_tenartsis','imc']
cat_cols = ['8_sexo','cigarrillo','20_dxdm','hta_tratada']  # binarios tratados como categóricos o numéricos; ambas opciones valen

pre = ColumnTransformer([
    ('num', 'passthrough', num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', drop=None), cat_cols),
])

base = LogisticRegression(max_iter=200, class_weight='balanced', solver='liblinear')

# Pipeline + calibración (Isotonic suele calibrar mejor)
clf = Pipeline([
    ('prep', pre),
    ('model', CalibratedClassifierCV(base, cv=5, method='isotonic'))
])

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
clf.fit(X_tr, y_tr)

# -------- 1) Evaluación --------
p_te = clf.predict_proba(X_te)[:,1]
print("AUROC:", roc_auc_score(y_te, p_te).round(3))
print("AUPRC:", average_precision_score(y_te, p_te).round(3))
print("Brier:", brier_score_loss(y_te, p_te).round(3))

# -------- 2) Prevalencia total y por segmento --------
def prevalencia(p):
    return float(np.nanmean(p))

prev_total = prevalencia(p_te)
print("Prevalencia estimada (test):", round(prev_total, 3))

# Por segmentos (ej. zona y comuna)
seg_cols = []
for col in ['Zona','COMUNA']:  # reemplaza con tus nombres exactos si están en X_te
    if col in df.columns:
        seg_cols.append(col)

if seg_cols:
    te_index = X_te.index
    df_seg = df.loc[te_index, seg_cols].copy()
    df_seg['p'] = p_te
    tabla = df_seg.groupby(seg_cols)['p'].mean().reset_index().rename(columns={'p':'prevalencia_estimada'})
    print(tabla.head())
```
